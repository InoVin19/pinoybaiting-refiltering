```python
merged_df = pd.read_csv('merged_output.csv').drop_duplicates()
merged_df
```

```python
#getting dataframe of videos with no transcripts
no_transcripts_df = merged_df[merged_df['Caption'] == False]
no_transcripts_df.shape[0]
```

```python
# Getting unique channel titles of videos with no transcripts
unique_channel_titles = no_transcripts_df['Channel Title'].unique()
channels_with_no_transcripts_df = pd.DataFrame(unique_channel_titles, columns=['Channel Title'])

# Adding Channel Id by merging with prev_df
channels_with_no_transcripts_df = channels_with_no_transcripts_df.merge(
    no_transcripts_df[['Channel Title', 'Channel Id']], on='Channel Title', how='left'
).drop_duplicates()

channels_with_no_transcripts_df
```

```python
# Adding column for number of videos per channel
channel_counts = no_transcripts_df['Channel Title'].value_counts()
channels_with_no_transcripts_df['videoCount'] = channels_with_no_transcripts_df['Channel Title'].map(channel_counts).fillna(0).astype(int)
channels_with_no_transcripts_df = channels_with_no_transcripts_df.sort_values(by='videoCount', ascending=False)
channels_with_no_transcripts_df
```

```python
prev_df = pd.read_csv('old_final_dataset.csv')
prev_df.shape[0]
```

```python
#removing rows from unique channels where channel exists in previous dataset
final_dataset_titles = prev_df['Channel Title'].unique()

channels_with_no_transcripts_df = channels_with_no_transcripts_df[~channels_with_no_transcripts_df['Channel Title'].isin(final_dataset_titles)]
channels_with_no_transcripts_df.shape[0]
```

```python
channels_temp = channels_with_no_transcripts_df[channels_with_no_transcripts_df['videoCount'] == 1]
channels_temp.shape[0]
```

```python
channels_with_no_transcripts_df.to_csv('channels_with_no_transcripts.csv', index=False)
```

```python
# Getting videos with no transcripts from filtered channels
filtered_no_transcripts_df = no_transcripts_df[no_transcripts_df["Channel Title"].isin(channels_with_no_transcripts_df["Channel Title"])]

filtered_no_transcripts_df['Channel Id'] = filtered_no_transcripts_df['Channel Id'].astype(str)
filtered_no_transcripts_df = filtered_no_transcripts_df.sort_values(by='Channel Id', ascending=True)

print(filtered_no_transcripts_df.shape[0])
```

```python
channel_id_to_title_map = channels_with_no_transcripts_df.set_index('Channel Id')['Channel Title'].to_dict()

filtered_no_transcripts_df = filtered_no_transcripts_df[[
    'Video Id', 'Video Title', 'Channel Id', 'Video Tags', 'Channel Country'
]].copy()

filtered_no_transcripts_df['Channel Title'] = filtered_no_transcripts_df['Channel Id'].map(channel_id_to_title_map)

filtered_no_transcripts_df = filtered_no_transcripts_df[
    ['Video Id', 'Video Title', 'Channel Id', 'Channel Title', 'Video Tags', 'Channel Country']
]

filtered_no_transcripts_df['Channel Id'] = filtered_no_transcripts_df['Channel Id'].astype(str)
filtered_no_transcripts_df = filtered_no_transcripts_df.sort_values(by='Channel Id', ascending=True).drop_duplicates()
```

```python
# ALL unique channels for filtering

unique_old_channels_df = prev_df[['Channel Title', 'Channel Id']].drop_duplicates()
unique_old_channels_df['videoCount'] = prev_df['Channel Title'].map(channel_counts).fillna(0).astype(int)

all_unique_channels_df = pd.concat([channels_with_no_transcripts_df, unique_old_channels_df])
all_unique_channels_df = all_unique_channels_df.drop_duplicates().sort_values(by='Channel Id', ascending=True)
all_unique_channels_df = all_unique_channels_df[all_unique_channels_df['videoCount'] >= 2]

all_unique_channels_df
```

```python
#Exporting all unique channels to csv
# all_unique_channels_df.to_csv('all_unique_channels.csv', index=False)
```

```python
# Getting all videos from previous dataset + videos with no transcripts from filtered channels
videos_for_filtering_df = merged_df[merged_df["Channel Title"].isin(all_unique_channels_df["Channel Title"])]

videos_for_filtering_df = videos_for_filtering_df.sort_values(by='Channel Id', ascending=True)

videos_for_filtering_df = videos_for_filtering_df.merge(
    prev_df[['Video Id', 'Video Title', 'Channel Id', 'Channel Title', 'Video Tags', 'Channel Country']], on='Video Id', how='left'
).drop_duplicates()

# Ensure unique Video Ids in merged_df before mapping
caption_map = merged_df.drop_duplicates(subset=['Video Id']).set_index('Video Id')['Caption']

# Map Caption values to videos_for_filtering_df
videos_for_filtering_df['Caption'] = videos_for_filtering_df['Video Id'].map(caption_map)

videos_for_filtering_df = videos_for_filtering_df[[
    'Video Id', 'Video Title_x', 'Channel Id_x', 'Channel Title_x', 'Video Tags_x', 'Channel Country_x'
]].copy()

videos_for_filtering_df
```
