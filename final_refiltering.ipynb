```python
import pandas as pd
import numpy as np
import os
import shutil
from langdetect import detect
pd.set_option('display.max_rows', 20)
```

```python
# Importing CSV of all annotations
annotated_df = pd.read_csv('annotated_csvs/annotated_channels.csv')

annotated_df
```

```python
# dropping rows with more "no" votes
filtered_channels = annotated_df[annotated_df["Pinoybaiting"] == "yes"]

filtered_channels.shape[0]
```

```python
# renaming old transcripts

for filename in os.listdir('2_Old_Transcripts'):
    if filename.endswith('_captions.txt'):
        new_filename = filename.replace("_captions.txt", ".txt")

        # Rename the file
        os.rename(os.path.join('old_transcripts', filename), os.path.join('old_transcripts', new_filename))
        print(f'Renamed: {filename} → {new_filename}')
```

```python
# filtering transcripts that are not in the filtered channels
all_videos = pd.read_csv('videos_for_filtering.csv')

filtered_videos = all_videos[all_videos['Channel Id_x'].isin(filtered_channels['Channel Id'])]
filtered_videos.drop_duplicates(subset='Video Id', inplace=True)
filtered_videos
```

```python
filtered_videos.to_csv('initial_filtering.csv', index=False)
```

```python
# removing files from directory
valid_video_ids = set(filtered_videos["Video Id"].astype(str))

for filename in os.listdir("3_Combined_Transcripts"):
    if filename.endswith(".txt"):
        video_id = filename.replace(".txt", "")  # Extract Video Id from filename

        # Check if the Video Id exists in filtered_videos
        if video_id not in valid_video_ids:
            file_path = os.path.join("3_Combined_Transcripts", filename)
            os.remove(file_path)  # Delete the file
            print(f"❌ Removed: {filename}")
```

```python
#filtering out videos that are not in English and have less than 100 words


directory = "3_Combined_Transcripts"
filtered_directory = "4_Final_Transcripts"

for filename in os.listdir(directory):
    if filename.endswith(".txt"):
        file_path = os.path.join(directory, filename)
        
        try:
            with open(file_path, "r", encoding="utf-8") as file:
                text = file.read()
            
            lang = detect(text)
            
            word_count = len(text.split())

            if lang == "en" and word_count >= 50:
                shutil.copy(file_path, os.path.join(filtered_directory, filename))
                print(f"✅ Kept: {filename} (Language: {lang}, Words: {word_count})")
            else:
                print(f"❌ Removed: {filename} (Language: {lang}, Words: {word_count})")

        except Exception as e:
            print(f"⚠️ Skipped {filename} due to error: {e}")

print("Filtering complete!")
```

```python
# updating csv with final filter
remaining_video_ids = {filename.replace(".txt", "") for filename in os.listdir('4_Final_Transcripts') if filename.endswith(".txt")}
filtered_videos = filtered_videos[filtered_videos["Video Id"].astype(str).isin(remaining_video_ids)]
filtered_videos.drop_duplicates(subset=['Video Id'], inplace=True)

filtered_videos
```

```python
filtered_videos.to_csv('final_filtering.csv', index=False)
```
